# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# Load Data and EDA

## Intraday 1 Min

```python
df = pd.read_csv("../datasets/intraday1min.csv", index_col=0)
print(df.shape)
df.head()
```

```python
df.index = pd.to_datetime(df.index)
df.columns = [x.split()[1] for x in df.columns]
print(type(df.index))
df.head()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
df.plot(y='close', ax=ax, lw=2)
plt.xlabel('Time')
plt.show()
```

```python
df.info()
```

Types check out, no need for changes here.

```python
df.describe()
```

This makes me want to drop the volume column

```python
df.isnull().sum()
```

Yeah! No nulls! This is expected but nice.
